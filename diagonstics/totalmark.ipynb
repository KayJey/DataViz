```python
data = pd.read_excel (r'TestAnalysis.xlsx')
print (data.columns)
```

```python
df = pd.DataFrame(data, columns=['Reg.No','Name','Scored'])
df
```

```python
df["Scored"].dtypes
```

```python
df["Scored"] = df["Scored"].astype(float)
```

```python
df = df[df['Scored'] != "-" ]  #to remove values that dont have a score
df["Scored"].dtypes 
```

```python
df["Scored"] = df["Scored"].astype(float)
df["Scored"].dtypes
```

```python
df = df[df["Scored"] >=20]
df.count #Reducing data to small number
```

```python
df = df[df["Scored"] >=26]
df.count()
#Reducing data to small number
```

```python
Scoredmarks = df["Scored"].values

Name = df['Name'].values

print(Name,Scoredmarks)
```

```python
%matplotlib inline
plt.figure()
yvals = Scoredmarks
xvals = range(len(Scoredmarks))
print(xvals,yvals)
```

```python
plt.bar(xvals , yvals, width=0.5 , color = 'green')
plt.xticks(xvals, Name)
plt.ylabel('% Scored')
x =plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
```

```python
df2 = pd.DataFrame(data, columns=['Reg.No','Name','Scored' ,'%'])
df2 = df2[df2['Scored'] != '-']
df2 = df2[df2['%'] != '-']
df2["Scored"] = df2["Scored"].astype(float)
df2["%"] = df2['%']*100

```
